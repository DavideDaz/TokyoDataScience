```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 500 # m

x_i =np.random.uniform(low=0, high=10, size=(observations,1))
x_i = np.sort(x_i, axis=0)
```

```python
noise = np.random.uniform(-1,1,(observations,1))
alfa_true = 2
beta_true =3
targets = alfa_true + x_i*beta_true + noise 
```

## Analytical Solution:


\begin{equation*}
\nabla L = \left ({\frac{\partial L}{\partial \alpha} , \frac{\partial L}{\partial \beta}} \right)=\left(\quad {-\frac{2}{m}\sum_{k=1}^n (y_i-(\alpha+\beta x_i))  , \quad -\frac{2}{m}\sum_{k=1}^n (x_i(y_i-(\alpha+\beta x_i)))} \right)
\end{equation*}

### Linear System:


<div style="font-size: 20px">
\begin{array}{c}
A=\sum_{k=1}^n y_i \quad
B=\sum_{k=1}^n x_i \quad
C=\sum_{k=1}^n x_iy_i \quad
D=\sum_{k=1}^n x_i^2 \quad\\
\\
\end{array}
<div>

<div style="font-size: 30px">
\begin{array}{c}
\beta & = \frac{A-m\alpha}{B} , \quad \alpha &= \frac{C-\left(1-\frac{A}{B}\right)}{B+\frac{Cm}{B}}
\end{array}

<div>

```python
A = sum(targets)
B = sum(x_i)
C = sum(x_i*targets)
D = sum(x_i*x_i)
```

```python
alfa = (C-D*A/B)/(B-observations*D/B)
beta = (A - alfa*observations)/B
print(alfa,beta)
```

## Gradient Descent Solution:


```python
init_range = 0.1

beta_gd = np.random.uniform(init_range, 2*init_range,size=1)
alfa_gd = np.random.uniform(init_range, 2*init_range,size=1)
                           
print(alfa_gd)
print(beta_gd)
```

```python
learning_rate = 0.0005
```

```python
for i in range(80000):
    outputs = (x_i*beta_gd) + alfa_gd
    deltas = outputs - targets
    
    loss = sum(deltas**2)/2/observations
    
    print(loss)
    
    deltas_scaled = deltas/ observations
    
    beta_gd = beta_gd - learning_rate * np.dot(x_i.T,deltas_scaled)
    alfa_gd = alfa_gd - learning_rate * np.sum(deltas_scaled)
```

```python
print(alfa_gd,beta_gd)
```

```python
y_analyt = alfa + beta*x_i

y_GD = alfa_gd + beta_gd*x_i
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x_i, targets, marker='o',color='black')
plt.plot(x_i,y_GD, color='red',linewidth=4)
plt.plot(x_i,y_analyt, color='green', Linestyle='--',linewidth=3)
plt.legend(( 'Gradient Descent','Analytical','Targets'),
           loc='lower right')
```
