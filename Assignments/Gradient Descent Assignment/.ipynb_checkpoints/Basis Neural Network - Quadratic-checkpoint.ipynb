```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 100 # m

x_i =np.random.uniform(low=0, high=10, size=(observations,1))
x_i = np.sort(x_i, axis=0)
```

```python
noise = np.random.uniform(-5,5,(observations,1))
alfa_true = 2
beta_true = 3
gamma_true = 8
targets = alfa_true + x_i*beta_true + gamma_true*x_i*x_i + noise 
```

## Analytical Solution:


\begin{equation*}
\nabla L = \left (\frac{\partial L}{\partial \alpha} , \frac{\partial L}{\partial \beta} , \frac{\partial L}{\partial \gamma} \right) =
\left(
\quad -\frac{2}{m}\sum_{k=1}^n (y_i-(\alpha+\beta x_i))  , 
\quad -\frac{2}{m}\sum_{k=1}^n (x_i(y_i-(\alpha+\beta x_i))) ,
\quad -\frac{2}{m}\sum_{k=1}^n (x_i^2(y_i-(\alpha+\beta x_i)))
\right)
\end{equation*}

```python
a = sum(targets)
b = sum(x_i)
c = sum(x_i*x_i)
d = sum(x_i*targets)
e = sum(x_i*x_i*x_i)
f = sum(x_i*x_i*targets)
g = sum(x_i*x_i*x_i*x_i)
m = observations
```

```python
alfa = -(b*(d*g-e*f)+c*(-a*g-e*d)+pow(c,2)*f+pow(e,2)*a)/((c*g-pow(e,2))*m-pow(b,2)*g-pow(c,3)+2*e*b*c)
beta = ((d*g-e*f)*m-a*b*g+c*(b*f+e*a)-pow(c,2)*d) /((c*g-pow(e,2))*m-pow(b,2)*g-pow(c,3)+2*e*b*c)
gamma = ((c*f-e*d)*m-pow(b,2)*f+b*c*d-a*pow(c,2)+e*a*b)/((c*g-pow(e,2))*m-pow(b,2)*g-pow(c,3)+2*e*b*c)
print(alfa,beta,gamma)
```

## Gradient Descent Solution:


```python
init_range = 0.1

beta_gd = np.random.uniform(init_range, 2*init_range,size=1)
alfa_gd = np.random.uniform(init_range, 2*init_range,size=1)
gamma_gd = np.random.uniform(init_range, 2*init_range,size=1)

print(alfa_gd)
print(beta_gd)
print(gamma_gd)
```

```python
learning_rate = 0.0005
```

```python
for i in range(80000):
    outputs = (x_i*x_i*gamma_gd) + (x_i*beta_gd) + alfa_gd
    deltas = outputs - targets
    
    loss = sum(deltas**2)/2/observations
    
    print(loss)
    
    deltas_scaled = deltas/ observations
    
    gamma_gd = gamma_gd - learning_rate * np.dot((x_i*x_i).T,deltas_scaled)
    beta_gd = beta_gd - learning_rate * np.dot(x_i.T,deltas_scaled)
    alfa_gd = alfa_gd - learning_rate * np.sum(deltas_scaled)
```

```python
print(alfa_gd,beta_gd, gamma_gd)
```

```python
y_analyt = alfa + beta*x_i + gamma*x_i*x_i

y_GD = alfa_gd + beta_gd*x_i + gamma_gd*x_i*x_i
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x_i, targets, marker='o', color='black' )
plt.plot(x_i,y_GD, color='red',linewidth=4)
plt.plot(x_i,y_analyt, color='green', Linestyle='--', linewidth=3)
plt.legend(( 'Gradient Descent','Analytical','Targets'),loc='lower right')
```
