```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 100 # m

x_i =np.random.uniform(low=0, high=10, size=(observations,1))
x_i = np.sort(x_i, axis=0)
```

```python
noise = np.random.uniform(-100,100,(observations,1))
alfa_true = 2
beta_true = 3
gamma_true = 8
targets = alfa_true + x_i*beta_true + gamma_true*x_i*x_i + noise 
```

## Gradient Descent Solution:


```python
loss_RMSProp = []

learning_rate = 0.01

r = np.zeros(3)

eps = 10**(-8)

rho = 0.9

Theta = np.array([0.17923125,0.10416243,0.12899188])
```

```python
for i in range(2000):
    
    #RSMProp
    
    outputs = (x_i*x_i*Theta[2]) + (x_i*Theta[1]) + Theta[0]
    deltas = outputs - targets
    
    loss = sum(deltas**2)/2/observations
    loss_RMSProp.append(loss[0])
    
    deltas_scaled = deltas/ observations
    
    g = np.array([np.sum(deltas_scaled),np.dot(x_i.T,deltas_scaled)[0][0],np.dot((x_i*x_i).T,deltas_scaled)[0][0]])
    
    r =np.add(np.multiply(r,rho),np.multiply((1-rho),np.multiply(g,g)))
    
    
    DeltaTheta = -np.multiply(np.divide(learning_rate,(eps + np.sqrt(r))),g)
    
    
    Theta += DeltaTheta
    
```

```python
print(Theta)
```

```python
y_RSMProp = Theta[0] + Theta[1]*x_i + Theta[2]*x_i*x_i
```

```python
%matplotlib inline
plt.scatter(x_i, targets, marker='o', color='black' )
plt.plot(x_i,y_RSMProp, color='yellow',linewidth=2)
plt.legend(('RSMProp','targets'),loc='lower right')
plt.xlabel('x')
plt.ylabel('y')
```

```python
%matplotlib inline
plt.plot(range(0,len(loss_RMSProp)),loss_RMSProp, color='magenta',linewidth=2)
plt.legend(('RSMProp',''),loc='upper right')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```
